```python
# Example of creating a white noise series.
```

```python
from random import gauss
from random import seed
from pandas import Series
from pandas.plotting import autocorrelation_plot
from matplotlib import pyplot
```

```python
# seed random number generator
seed(1)
```

```python
# create white noise series
series = [gauss(0.0, 1.0) for i in range(1000)]
series = Series(series)
```

```python
# summary stats
print(series.describe())
```

```python
# We can see that the mean is nearly 0.0 and the standard deviation is nearly 1.0. Some
# variance is expected given the small size of the sample.
```

```python
# line plot
series.plot()
pyplot.show()
```

```python
# histogram plot
series.hist()
pyplot.show()
```

```python
# Indeed, the histogram shows the bell-curve shape.
```

```python
# autocorrelation
autocorrelation_plot(series)
pyplot.show()
```

```python
# The correlogram does not show any obvious autocorrelation pattern. There are some spikes
# above the 95% and 99% confidence level, but these are a statistical 
# fluke.
```
